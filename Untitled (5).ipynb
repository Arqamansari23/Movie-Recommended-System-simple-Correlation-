# Recommending similar movies 

Now let's create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies.

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

### finding most rated movie

```python
ratings.sort_values(by='number of ratings',ascending=False).head(10)
```

stars wars is the most rated movie  

###  Now let's grab the user ratings for Star wars movie:

```python
starwars_user_ratings = moviemat['Star Wars (1977)']  # all the ratings by the users given to stars wars
```

```python
starwars_user_ratings
```

### We can then use corrwith() method to get correlations between two pandas series:

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)  # the stars wars movie is corelated with all the outher movies to find best corelated values 
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation']) #  This data frame is of correlation of movies with stars wars 
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

Now if we sort the dataframe by correlation, we should get the most similar movies, however note that we get some results that don't really make sense. This is because there are a lot of movies only watched once by users who also watched star wars (it was the most popular movie).

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

 this is possible that only one or 2 user gave them 5 ratings so there corelation is 1 
 
 this problem can be removed by if we take only that movies which review count is greater than 100

```python
corr_starwars = corr_starwars.join(ratings['number of ratings']) # We combine count of ratings with corelation 
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```
